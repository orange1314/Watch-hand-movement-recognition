```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

# 定義資料夾路徑
a_folder = 'data/Game'
b_folder = 'test/Game'

# 確保 b 資料夾存在
os.makedirs(b_folder, exist_ok=True)

# 遍歷 a 資料夾中的所有 CSV 文件
for filename in os.listdir(a_folder):
    if filename.endswith('.csv'):
        # 讀取 CSV 文件
        file_path = os.path.join(a_folder, filename)
        df = pd.read_csv(file_path, encoding='ISO-8859-1')
        
        # 檢查資料是否為空
        if df.empty:
            print(f"{filename} is empty. Skipping this file.")
            continue

        # 分割資料集，70% 保留在原始檔案，30% 存儲到 b 資料夾
        train, test = train_test_split(df, test_size=0.3, random_state=42)

        # 將 70% 資料保存到原始檔案中
        train.to_csv(file_path, index=False)

        # 將 30% 資料保存到 b 資料夾
        test_path = os.path.join(b_folder, filename)
        test.to_csv(test_path, index=False)

print("資料分割完成。")
```
